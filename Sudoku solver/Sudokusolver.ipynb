```python
import numpy as np
```

```python
class sudoku_solver:
    # find an element
    # try a value
    def __init__(self, List):
        # initialize with a list of tuples where each tuple consists of (value, row, col)
        self.board = np.zeros([9, 9])
        for tup in List:
            self.board[tup[1], tup[2]] = tup[0]
    def disp(self):
        print(np.matrix(self.board))

    def find_box(self, position):
        # given position, return elemenets in the box containing [position]
        i = position[0]//3 * 3
        j = position[1]//3 * 3
        return self.board[i:i + 3, j:j + 3]
        
    def is_valid(self, number, position):
        if number not in self.board[position[0],:]:
            if number not in self.board[:, position[1]]:
                if number not in self.find_box(position):
                    return True
        return False
        
    def solve(self):
        for i in range(9):
            for j in range(9):
                if self.board[i][j] == 0:
                    for n in range(1, 10):
                        if self.is_valid(n, (i, j)):
                            self.board[i][j] = n
                            self.solve()
                            self.board[i][j] = 0
                    return
        self.disp()
        input('?')

        
```

```python
board = zip([7, 3, 2, 6, 4, 8, 3, 5, 7, 1, 8, 4, 2, 3, 7, 9, 6, 3, 1, 5],#, 4, 2, 2, 9, 6],
           [0, 0, 0, 1, 1, 2, 2, 2, 2, 3, 3, 3, 4, 5, 5, 5, 6, 6, 6, 6],#, 7, 7, 8, 8, 8], 
           [2,3 , 5, 2, 6, 0, 1, 3, 7, 0, 2, 8, 4, 0, 6, 8, 1, 5, 7, 8])#, 2, 6, 3, 5, 6])

sudoku = sudoku_solver(board)
sudoku.disp()
```
